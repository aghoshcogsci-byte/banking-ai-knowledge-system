```python
def validate_intent_triggers():

    query = """
    MATCH (i:Intent)
    WHERE NOT (i)-[:TRIGGERS]->(:ServiceFlow)
    RETURN i.name AS invalid_intent
    """

    with driver.session() as session:
        result = session.run(query)
        return [r["invalid_intent"] for r in result]
```

```python
def validate_service_verification():

    query = """
    MATCH (s:ServiceFlow)
    WHERE NOT (s)-[:REQUIRES]->(:Verification)
    RETURN s.name AS invalid_service
    """

    with driver.session() as session:
        result = session.run(query)
        return [r["invalid_service"] for r in result]
```
